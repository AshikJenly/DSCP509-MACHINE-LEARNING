```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dropout,Flatten,MaxPool2D,InputLayer,Dense
from tensorflow.keras.losses import binary_crossentropy
```

```python
X_train_org, y_train, X_test_org, y_test = np.load('datas/Files/cat_train_x.npy'),np.load('datas/Files/cat_train_y.npy'),np.load('datas/Files/cat_test_x.npy'),np.load('datas/Files/cat_test_y.npy')
train_x_flatten = X_train_org.reshape(X_train_org.shape[0], -1).T   # The "-1" makes reshape flatten the remaining dimensions
test_x_flatten = X_test_org.reshape(X_test_org.shape[0], -1).T

X_train = train_x_flatten/255.
X_test = test_x_flatten/255.
```

```python
model = Sequential([
    Dense(20,activation="relu",input_shape=(1,12288)),
    Dense(7,activation="relu"),
    Dense(5,activation="relu"),
    Dense(1,activation="sigmoid")
]
)
```

```python
model.compile( loss=binary_crossentropy,metrics=["accuracy"])
```

```python
model.summary()
```

```python
x_train_tensor = tf.convert_to_tensor(np.expand_dims(X_train.T,1))
y_train_tensor = tf.convert_to_tensor(y_train.T)
```

```python
model.fit(x=x_train_tensor,y=y_train_tensor,epochs=2400)
```

```python
predictions = model.predict(tf.convert_to_tensor(np.expand_dims(X_test.T,1)))
```

```python
predicted_labels = (predictions >= 0.5).astype(int)
```

```python
predicted_labels = predicted_labels.reshape(1, 50)
```

```python
correctly_identified = np.sum(predicted_labels == y_test)
```
